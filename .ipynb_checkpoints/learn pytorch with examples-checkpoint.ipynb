```python
import torch
import numpy as np
```

```python
N,D_in,H,D_out=64,1000,100,10
x=np.random.randn(N,D_in)
y=np.random.randn(N,D_out)

w1=np.random.randn(D_in,H)
w2=np.random.randn(H,D_out)

learning_rate=1e-6
```

```python
for t in range(500):
    #Forward
    h=x.dot(w1)   #(N,H)
    h_relu=np.maximum(h,0)
    y_pred=h_relu.dot(w2) #(N,D_out)
    
    #计算loss
    loss=np.square(y_pred-y).sum()
    print(t,loss)
    
    #BP
    grad_y_pred=2*(y_pred-y)   #（N,D_out）
    grad_w2=h_relu.T.dot(grad_y_pred)  #(H,N)  (N,D_out)
    grad_h_relu=grad_y_pred.dot(w2.T)
    grad_h=grad_h_relu.copy()
    grad_h[h<0]=0
    grad_w1=x.T.dot(grad_h)
    
    w1-=learning_rate*grad_w1
    w2-=learning_rate*grad_w2
```

```python
class MyRelu(torch.autograd.Function):
    @staticmethod
    def forward(ctx,input):
        '''
        前向传播接受一个Tensor并返回一个Tensor.
        ctx是一个上下文对象能够用来存储用于反向传播的梯度信息，
        通过ctx.svae_for_backward存储
        '''
        ctx.save_for_backward(input)
        return input.clamp(min=0)
    @staticmethod
    def backward(ctx,grad_output):
        '''
        BP环节，接收了损失函数关于参数的梯度
        '''
        input,=ctx.saved_tensors
        grad_input=grad_output.clone()
        grad_input[input<0]=0
        return grad_input
device=torch.device('cpu')

N,D_in,H,D_out=64,1000,100,10

x=torch.randn(N,D_in,device=device,dtype=torch.float)
y=torch.randn(N,D_out,device=device,dtype=torch.float)

w1=torch.randn(D_in,H,device=device,dtype=torch.float,requires_grad=True)
w2=torch.randn(H,D_out,device=device,dtype=torch.float,requires_grad=True)

learning_rate=1e-6
for i in range(500):
    relu=MyRelu.apply
    y_pred=relu(x.mm(w1)).mm(w2) 
    loss=(y_pred-y).pow(2).sum()
    
    if t%100==99:
        print(t,loss.item())
    loss.backward()
    with torch.no_grad():
        w1-=learning_rate*w1.grad
        w2-=learning_rate*w2.grad
        
        w1.grad.zero_()
        w2.grad.zero_()
```

```python
##nn package实现
#device=torch.device('cuda')
x=torch.randn(N,D_in,device='cpu')
y=torch.randn(N,D_out,device='cpu')

model=torch.nn.Sequential(torch.nn.Linear(D_in,H),
                         torch.nn.ReLU(),
                        torch.nn.Linear(H,D_out))
#reduce决定返回的loss是一个标量还是tensor，为True时默认为标量
#size_average在reduce=True时起作用，size_average=True返回的是样本的平均Loss,否则返回的时所有样本loss的和
loss_fn=torch.nn.MSELoss(reduce=True,size_average=True)
learning_rate=1e-4
for i in range(5000):
    y_pred=model(x)
    loss=loss_fn(y_pred,y)
    
    if i%100==0:
        print(i,loss.item())
        
    model.zero_grad()
    
    loss.backward()
    
    with torch.no_grad():
        for param in model.parameters():
            param-=learning_rate*param.grad
```

```python
#定义自定义网络模型
import torch
import torch.nn.functional as F
```

```python
class TwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super(TwoLayerNet,self).__init__()
        self.linear1=torch.nn.Linear(D_in,H)
        self.linear2=torch.nn.Linear(H,D_out)
    def forward(self,x):
        x=self.linear1(x)
        h_relu=F.relu(x)
        y_pred=self.linear2(h_relu)
        return y_pred
    
        
N,D_in,H,D_out=64,1000,100,10
x=torch.randn(N,D_in)
y=torch.randn(N,D_out)

model=TwoLayerNet(D_in,H,D_out)

criterion=torch.nn.MSELoss(reduction='sum')
optimizer=torch.optim.SGD(model.parameters(),lr=1e-4)

for i in range(500):
    y_pred=model(x)
    loss=criterion(y_pred,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if i%50==0:
        print(i,loss.item())
```

```python
#Control Flow+Wright Sharing
#不同于Lua torch，只能将一个Module使用一次，如今的torch在定义一个动态计算图时可以重复多次使用
#同一个Module，并可以做到权重共享
import random
```

```python
class DynamicNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super(DynamicNet,self).__init__()
        self.input_linear=torch.nn.Linear(D_in,H)
        self.middle_linear=torch.nn.Linear(H,H)
        self.output_linear=torch.nn.Linear(H,D_out)
    def forward(self,x):
        x=self.input_linear(x)
        h_relu=F.relu(x)
        for _ in range(random.randint(0,3)):
            h_relu=self.middle_linear(h_relu).clamp(min=0)
        y_pred=self.output_linear(h_relu)
        return y_pred
model=DynamicNet(D_in,H,D_out)
print(model)
g=model.parameters()
g=list(g)
len(g)
print(g[0].shape)
print(g[1].shape)
print(g[2].shape)
print(g[3].shape)
```
